```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

import numpy as np


RS = np.arange(0,50)
Training_score = []
Testing_score = []

for random_state in RS:
    
    X_train, X_test, y_train, y_test = train_test_split(X_Transformed_PF,\
Boston_data.target, random_state = random_state)
    LRObject = LinearRegression()
    LRFitting = LRObject.fit(X_train, y_train)
    Training_score.append(LRFitting.score(X_train, y_train))
    Testing_score.append(LRFitting.score(X_test, y_test))
    
plt.figure(figsize=(10,8))
plt.plot(RS, Training_score, c='red', label = 'Training score')
plt.plot(RS, Testing_score, c='blue', label = 'Testing score')

plt.xlabel('random_states')
plt.ylabel('Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures

Boston_data = load_boston()
MMObject = MinMaxScaler()
X_Transformed = MMObject.fit_transform(Boston_data.data)

PFObject = PolynomialFeatures(2)
X_Transformed_PF = PFObject.fit_transform(X_Transformed)

X_train, X_test, y_train, y_test = train_test_split(X_Transformed_PF, Boston_data.target, random_state = 10)
Alpha = np.linspace(0.001,0.3,40)

Training_score = []
Testing_score = []
Number_of_Features = []

for alpha in Alpha:
    LASSO_Object = Lasso(alpha = alpha, max_iter=100000)
    LAFitting = LASSO_Object.fit(X_train, y_train)
    Training_score.append(LAFitting.score(X_train, y_train))
    Testing_score.append(LAFitting.score(X_test, y_test))
    Number_of_Features.append(np.sum(LAFitting.coef_!=0))

import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.plot(Alpha, Training_score, c='red', label = 'Training score')
plt.plot(Alpha, Testing_score, c='blue', label = 'Testing score')
plt.legend()
plt.xlabel('alpha')
plt.ylabel('score')

plt.figure(figsize=(10,8))
plt.plot(Alpha, Number_of_Features, c='red')
plt.xlabel('alpha')
plt.ylabel('Number of nonzero features')

plt.show()
```
