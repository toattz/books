```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
data.data[0:2]
```

```python
data.target
```

```python
data.feature_names
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNNObject = KNeighborsClassifier(n_neighbors = 3)
```

```python
KNNFitting = KNNObject.fit(X_train,y_train)
```

```python
KNNFitting.score(X_train,y_train)
```

```python
KNNFitting.score(X_test,y_test)
```

```python
import numpy as np

no_numbers = np.arange(1,13,2)

Training_accuracy = []
Testing_accuracy = []

for no in no_numbers:
    KNNObject = KNeighborsClassifier(n_neighbors = no)
    KNNFitting = KNNObject.fit(X_train,y_train)
    Training_accuracy.append(KNNFitting.score(X_train,y_train))
    Testing_accuracy.append(KNNFitting.score(X_test,y_test))
```

```python
Training_accuracy
```

```python
Testing_accuracy
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[8,6])

plt.plot(no_numbers,Training_accuracy, label = 'Training Accuracy')
plt.plot(no_numbers,Testing_accuracy, label = 'Testing Accuracy')

plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=40)

no_numbers = np.arange(1,13,2)

Training_accuracy = []
Testing_accuracy = []

for no in no_numbers:
    KNNObject = KNeighborsClassifier(n_neighbors = no)
    KNNFitting = KNNObject.fit(X_train,y_train)
    Training_accuracy.append(KNNFitting.score(X_train,y_train))
    Testing_accuracy.append(KNNFitting.score(X_test,y_test))
    
plt.figure(figsize=[8,6])

plt.plot(no_numbers,Training_accuracy, label = 'Training Accuracy')
plt.plot(no_numbers,Testing_accuracy, label = 'Testing Accuracy')

plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.legend()
```
