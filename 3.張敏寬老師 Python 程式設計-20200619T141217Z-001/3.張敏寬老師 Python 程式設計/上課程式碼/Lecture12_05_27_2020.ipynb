```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

Accuracy_training = []
Accuracy_testing = []

No_Neighbors = np.arange(1,13,2)

for no_neighbor in No_Neighbors:
    KNN_Object = KNeighborsClassifier(n_neighbors = no_neighbor)
    KNN_Object.fit(X_train, y_train)
    Accuracy_training.append(KNN_Object.score(X_train, y_train))
    Accuracy_testing.append(KNN_Object.score(X_test, y_test))
```

```python
Accuracy_training
```

```python
Accuracy_testing
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.plot(No_Neighbors, Accuracy_training, c='red', label = 'Training accuracy')
plt.plot(No_Neighbors, Accuracy_testing, c='blue', label = 'Testing accuracy')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_samples = 60

uniform_x = np.random.uniform(-3, 3, n_samples)

y_true = np.sin(4*uniform_x) + uniform_x

y = (y_true + np.random.normal(0,1,n_samples))/2

plt.scatter(uniform_x, y_true, c = 'blue')
plt.scatter(uniform_x, y, c = 'red')
plt.show()
```

```python
uniform_x.shape
```

```python
from sklearn.model_selection import train_test_split
X = uniform_x.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0)
```

```python
from sklearn.linear_model import LinearRegression

LRObject = LinearRegression()
```

```python
LRFitting = LRObject.fit(X_train, y_train)
```

```python
LRFitting.predict(X_test)
```

```python
y_test
```

```python
LRFitting.score(X_test, y_test)
```

```python
y_diff = LRFitting.predict(X_test) - y_test
```

```python
y_diff.shape
```

```python
y_diff_square = 0
for i in y_diff:
    y_diff_square = i*i + y_diff_square
```

```python
y_diff_square
```

```python
y_diff_square/15
```

```python
LRFitting.coef_
```

```python
LRFitting.intercept_
```

```python
y_predit = LRFitting.coef_[0]*X_test + LRFitting.intercept_
```

```python
y_predit
```

```python
LRFitting.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_min, x_max = X[:,0].min() -1, X[:,0].max() + 1

Lines = np.linspace(x_min, x_max, 100).reshape(-1,1)

Line_predict = LRFitting.predict(Lines)

plt.figure(figsize=(10,8))

plt.plot(Lines, Line_predict, c = 'blue')
plt.scatter(uniform_x, y, c = 'red')

plt.show()
```
