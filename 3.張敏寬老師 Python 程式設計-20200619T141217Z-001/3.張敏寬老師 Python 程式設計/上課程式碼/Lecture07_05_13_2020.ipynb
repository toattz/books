```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,2)

ax01 = plt.subplot(2,2,1)
x = [1, 2, 3]
y = [8, 9, 10]
plt.plot(x,y)
ax01.set_title('Index 1')

ax02 = plt.subplot(2,2,2)
x = [1, 4, 3]
y = [8, 9, 10]
plt.plot(x,y)
ax02.set_title('Index 2')

ax03 = plt.subplot(2,2,3)
x = [1, 2, 5]
y = [8, 9, 10]
plt.plot(x,y)
ax03.set_title('Index 3')

ax04 = plt.subplot(2,2,4)
x = [1, 3, 5]
y = [8, 9, 10]
plt.plot(x,y)
ax04.set_title('Index 4')

plt.show()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots()
x = [1, 4, 3]
y = [8, 9, 10]
plt.plot(x,y, label = 'Legend')
axes.legend(loc = 'lower right')
axes.set_xlabel('X')
axes.set_ylabel('Y')
fig.savefig('test.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots()
x = np.arange(0,20,0.01)
y1 = np.sin(x)
y2 = np.cos(x)

axes.plot(x,y1,label = 'sin')
axes.plot(x,y2,label = 'cos')
axes.legend(loc = 'upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots()
x = [1, 2, 3]
y = [6, 3, 10]
labels = ['a', 'b', 'c']
axes.bar(x,y,tick_label = labels)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots()
x = [1, 2, 3]
y = [6, 3, 10]
labels = ['a', 'b', 'c']
axes.barh(x,y,tick_label = labels)
plt.show()
```
