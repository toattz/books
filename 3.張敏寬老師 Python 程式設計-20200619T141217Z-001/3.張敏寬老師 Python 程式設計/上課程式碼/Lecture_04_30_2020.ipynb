```python
Grade = int(input('成績為: '))
if Grade>=90:
    print(f'成績為 {Grade},等第為優等')
elif 80<=Grade<90:
    print(f'成績為 {Grade},等第為甲等')
elif 70<=Grade<80:
    print(f'成績為 {Grade},等第為乙等')
elif 60<=Grade<70:
    print(f'成績為 {Grade},等第為丙等')
else:
    print(f'成績為 {Grade},等第為不及格')
```

```python
count = 1;
one_pre = 1;
two_pre = 1;

while count<=18:
    current = one_pre + two_pre;
    two_pre = one_pre
    one_pre = current
    count = count + 1
    print(current, end=',')
    
```

```python
count = 1;
List_result = [1, 1];

while count<=18:
    current = List_result[0] + List_result[1]
    List_result[:0] = [current]
    count = count + 1

List_result.sort()
print(List_result)
```

```python
count = 1;
List_result = [1, 1];

while count<=18:
    current = List_result[-1] + List_result[-2]
    List_result[len(List_result):] = [current]
    count = count + 1

List_result.sort()
print(List_result)
```

```python
List = [1, 2, 3, 4, 5, 6, 7, 8, 9]
count = -1
while True:
    count = count + 1
    if List[count] == 6:
        continue
    if List[count] == 8:
        break
    print(List[count])
    
```
