```python
from sklearn.datasets import load_breast_cancer

BC = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(BC.data, BC.target, random_state = 0)
```

```python
from sklearn.svm import LinearSVC

LSVCObject = LinearSVC(C=1, max_iter=100000)
```

```python
LSVCFitting = LSVCObject.fit(X_train, y_train)
```

```python
LSVCFitting.score(X_train, y_train)
```

```python
LSVCFitting.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

C_value = np.array([1, 2, 4, 6, 8, 10, 15, 20])
Score_train = []
Score_test = []

for c in C_value:
    LSVCObject = LinearSVC(C=c, max_iter=200000)
    LSVCFitting = LSVCObject.fit(X_train, y_train)
    Score_train.append(LSVCFitting.score(X_train, y_train))
    Score_test.append(LSVCFitting.score(X_test, y_test))

plt.figure(figsize=(10,8))
plt.plot(C_value, Score_train, c='red', label = 'Training accuracy')
plt.plot(C_value, Score_test, c='blue', label = 'Testing accuracy')

plt.xlabel('C values')
plt.ylabel('Accuracy')
plt.legend(loc = 'upper right')

plt.show()
```

```python
from sklearn.datasets import load_digits

Digits_data = load_digits()
```

```python
Digits_data.data[0]
```

```python
Digits_data.target[0]
```

```python
Digits_data.target_names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Digits_data.data, Digits_data.target, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

C_value = np.array([1, 2, 4, 6, 8, 10, 15, 20])

LR_Training_Score = []
LR_Test_Score = []
LSVC_Training_Score = []
LSVC_Test_Score = []

for c in C_value:
    LRObject = LogisticRegression(C=c , max_iter = 10000, solver = 'lbfgs')
    LSVCObject = LinearSVC(C=c, max_iter = 10000)
    
    LRFitting = LRObject.fit(X_train, y_train)
    LSVCFitting = LSVCObject.fit(X_train, y_train)
    
    LR_Training_Score.append(LRFitting.score(X_train,y_train))
    LSVC_Training_Score.append(LSVCFitting.score(X_train,y_train))
    
    LR_Test_Score.append(LRFitting.score(X_test,y_test))
    LSVC_Test_Score.append(LSVCFitting.score(X_test,y_test))

plt.figure(figsize=(10,8))

plt.plot(C_value, LR_Training_Score, c='red', label = 'LogisticRegression Training Score')
plt.plot(C_value, LR_Test_Score, c='black', label = 'LogisticRegression Test Score')

plt.plot(C_value, LSVC_Training_Score, c='blue', label = 'LinearSVC Training Score')
plt.plot(C_value, LSVC_Test_Score, c='green', label = 'LinearSVC Test Score')

plt.xlabel('C values')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
X, y = make_moons(noise=0.2, random_state=2)

plt.figure(figsize=(10,8))

plt.scatter(X[y==0,0],X[y==0,1], c='blue', marker = 'o', label= 'Class 0')
plt.scatter(X[y==1,0],X[y==1,1], c='red', marker = '^', label= 'Class 1')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
DTCObject = DecisionTreeClassifier(max_depth=3, random_state=0)
DTCFitting = DTCObject.fit(X,y)
```

```python
DTCObject.tree_.threshold
```

```python
DTCObject.tree_.children_left
```

```python
DTCObject.tree_.children_right
```

```python
DTCObject.tree_.node_count
```

```python
import numpy as np

depth_information = np.zeros(DTCObject.tree_.node_count)
parent_information = np.zeros(DTCObject.tree_.node_count)

for nodes in np.arange(DTCObject.tree_.node_count):
    if DTCObject.tree_.children_left[nodes]!=DTCObject.tree_.children_right[nodes]:
        print(f'Node {nodes} in the decision tree has its left child: Node {DTCObject.tree_.children_left[nodes]}')
        print(f'Node {nodes} in the decision tree has its right child: Node {DTCObject.tree_.children_right[nodes]}')
        
        depth_information[DTCObject.tree_.children_left[nodes]] = depth_information[nodes]+1
        depth_information[DTCObject.tree_.children_right[nodes]] = depth_information[nodes]+1
        parent_information[DTCObject.tree_.children_left[nodes]] = nodes
        parent_information[DTCObject.tree_.children_right[nodes]] = nodes
        
        print(f'Current depth is {depth_information[nodes]}')
        print(f'The threshold at Node {nodes} in the decision tree is {DTCObject.tree_.threshold[nodes]}\n')
```
