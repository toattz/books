```python
def login_func(username, password):
    personal_info = open('user_password.txt','r')
    user_password = personal_info.read()
    user_password_dict = ast.literal_eval(user_password)
    personal_info.close()
    if user_password_dict[username] == password: 
        return True
    else:
        return False
```

```python
username = input('請輸入帳號: ')
password = input('請輸入密碼: ')
Boolvalue = login_func(username,password)
if Boolvalue == True:
    print('correct!!')
else:
    print('incorrect!!')
```

```python
def change_password(username):
    personal_info = open('user_password.txt','r')
    user_password = personal_info.read()
    user_password_dict = ast.literal_eval(user_password)
    personal_info.close()
    new_password = input('請輸入新的密碼: ')
    user_password_dict[username] = new_password
    personal_info = open('user_password.txt','w+')
    personal_info.write(str(user_password_dict))
    personal_info.close()
```

```python
username = 'John'
change_password(username)
```

```python
def Add_new_user(username,password):
    personal_info = open('user_password.txt','r')
    user_password = personal_info.read()
    user_password_dict = ast.literal_eval(user_password)
    personal_info.close()
    user_password_dict[username] = password
    personal_info = open('user_password.txt','w+')
    personal_info.write(str(user_password_dict))
    personal_info.close()    
```

```python
Add_new_user('Bill','888')
```

```python
def Delete_user(username):
    personal_info = open('user_password.txt','r')
    user_password = personal_info.read()
    user_password_dict = ast.literal_eval(user_password)
    personal_info.close()
    personal_info = open('user_password.txt','w+')
    user_password_dict.pop(username)
    personal_info.write(str(user_password_dict))
    personal_info.close()    
```

```python
Delete_user('Bill')
```

```python
while True:
    print('選項1: 登入系統: ')
    print('選項2: 新增使用者: ')
    print('選項3: 刪除使用者: ')
    print('選項4: 離開系統: ')
    Option = input('請輸入選項: ')
    if Option == '1':
        username = input('請輸入帳號: ')
        password = input('請輸入密碼: ')
        Boolvalue = login_func(username,password)
        if Boolvalue == True:
            Option2 = input('請問要更改密碼嗎?(Y/N) ')
            if Option2 == 'Y':
                change_password(username)
                Option = '4'
            elif Option2 == 'N':
                Option = '4'
        else:
            print('輸入密碼錯誤')
            Option = '4'
    if Option == '2':
        username = input('請輸入新的帳號: ')
        password = input('請輸入新的密碼: ')
        Add_new_user(username,password)
        Option = '4'
            
    if Option == '3':
        username = input('請輸入要刪除的帳號: ')
        Delete_user(username)
        Option = '4'
        
    if Option == '4':
        break;
        
```
