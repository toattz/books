```python
def converter(degree, Operation = 'F'):
    if Operation == 'F':
        Output = degree*9/5+32
        print(f'攝氏 {degree} 度等於華氏 {Output}')
    if Operation == 'C':
        Output = (degree-32)*5/9
        print(f'華氏 {degree} 度等於攝氏 {Output}')
    
```

```python
converter(32)
```

```python
converter(32,'C')
```

```python
def multiple_function(*a,**b):
    if b['op1']==1:
        result_op1 = []
        for i in range(len(a)-1):
            result_op1.append(a[i+1]-a[i])
        print(f'數列的鄰近差為 {result_op1}')
        
    if b['op2']==1:
        result_op2 = []
        for i in range(len(a)-1):
            result_op2.append(a[i+1]+a[i])
        print(f'數列的鄰近和為 {result_op2}')
            
    if b['op3']==1:
        result_op3 = []
        for i in range(len(a)-1):
            result_op3.append(a[i+1]*a[i])
        print(f'數列的鄰近乘積為 {result_op3}')
```

```python
multiple_function(1,2,3,4,op1=1,op2=0,op3=1)
```
