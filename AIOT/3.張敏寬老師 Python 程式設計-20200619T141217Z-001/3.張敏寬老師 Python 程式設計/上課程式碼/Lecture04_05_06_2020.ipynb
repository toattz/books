```python
def guess_num():
    default_num = '122'
    while True:
        count01 = 0;
        count02 = 0;
        your_guess = input('請輸入你猜的數字: ')
        if your_guess == default_num:
            print('恭喜你答對了!!!')
            break
        else:
            for i in range(len(default_num)):
                for j in range(len(your_guess)):
                    if i==j:
                        if default_num[i]==your_guess[j]:
                            count02 = count02 + 1
                    else:
                        if default_num[i]==your_guess[j]:
                            count01 = count01 + 1
        print(f'{count02}A{count01}B')
```

```python
guess_num()
```
