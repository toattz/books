```python
from tensorflow.keras.datasets import mnist
(train_X, train_y), (test_X, test_Y) = mnist.load_data()
```

```python
from tensorflow.keras.utils import to_categorical

X_train = train_X.reshape(60000,28*28)
X_test = test_X.reshape(test_X.shape[0],test_X.shape[1]*test_X.shape[2])
X_train = X_train.astype('float')/255
X_test = X_test.astype('float')/255

y_train = to_categorical(train_y)
y_test = to_categorical(test_Y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


Model07 = Sequential()
Model07.add(Dense(512, activation = 'relu', input_dim = 28*28))
Model07.add(Dense(10, activation = 'softmax'))
Model07.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
Model07.fit(X_train, y_train, epochs = 5, batch_size = 128)
```

```python
Model_weight = Model07.get_weights()
```

```python
Model_weight
```

```python
Model07.save_weights('Model_weight')
```

```python
Model07_config = Model07.get_config()
```

```python
from tensorflow.keras.models import Sequential

Model08 = Sequential.from_config(Model07_config)
```

```python
Model08.summary()
```

```python
Model08.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
Model08.set_weights(Model_weight)
```

```python
Model08.predict_classes(X_test)
```

```python
Model08.load_weights('Model_weight')
```

```python
Model08.predict_classes(X_test)
```

```python
from tensorflow.keras.models import Sequential

Model09 = Sequential.from_config(Model07_config)
```

```python
from tensorflow.keras import optimizers

sgd_new = optimizers.SGD(learning_rate = 0.5, momentum = 0.9)

Model08.compile(loss = 'categorical_crossentropy', optimizer = sgd_new, metrics = ['accuracy'])
```

```python
Model08.fit(X_train, y_train, epochs = 5, batch_size = 128)
```

```python
Model08.evaluate(X_test, y_test)
```

```python
from tensorflow.keras.datasets import imdb

(train_X, train_y), (test_X, test_y) = imdb.load_data(num_words = 1200)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

TKObject = Tokenizer(num_words = 1200)
```

```python
X_train = TKObject.sequences_to_matrix(train_X)
```

```python
X_test = TKObject.sequences_to_matrix(test_X)
```

```python
y_train = train_y.astype('float')
y_test = test_y.astype('float')
```

```python
from tensorflow.keras.models import Sequential

Model09 = Sequential()
```

```python
from tensorflow.keras.layers import Dense

Model09.add(Dense(32, activation = 'relu', input_dim = 1200))
Model09.add(Dense(16, activation = 'relu'))
Model09.add(Dense(8, activation = 'relu'))
Model09.add(Dense(1, activation = 'sigmoid'))
```

```python
Model09.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = 'accuracy')
```

```python
Model09.fit(X_train, y_train, epochs = 10, batch_size = 128, validation_split = 0.1)
```
