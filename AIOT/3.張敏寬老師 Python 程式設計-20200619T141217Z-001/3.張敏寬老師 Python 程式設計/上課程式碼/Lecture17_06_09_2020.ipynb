```python
from tensorflow.keras.datasets import mnist
(train_X, train_y), (test_X, test_Y) = mnist.load_data()
```

```python
train_X.shape
```

```python
test_X.shape
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,5,figsize = (10,8))

for axe, i in zip(axes.ravel(),range(5)):
    axe.imshow(train_X[i], cmap = 'gray')
    axe.set_title(f'Label = {train_y[i]}')
    
plt.show()
```

```python
X_train = train_X.reshape(60000,28*28)
```

```python
train_X[0]
```

```python
X_train[0]
```

```python
X_train[0].shape
```

```python
train_X[0].shape
```

```python
X_test = test_X.reshape(test_X.shape[0],test_X.shape[1]*test_X.shape[2])
```

```python
test_X.shape
```

```python
X_test.shape
```

```python
X_train = X_train.astype('float')
```

```python
X_train[0][0:5]
```

```python
X_train = X_train/255
```

```python
X_test = X_test.astype('float')/255
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(train_y)
```

```python
y_train[0:2]
```

```python
train_y[0:2]
```

```python
y_test = to_categorical(test_Y)
```

```python
from tensorflow.keras.models import Sequential
Model = Sequential()
```

```python
from tensorflow.keras.layers import Dense

Model.add(Dense(512, activation = 'relu', input_dim = 28*28))
```

```python
Model.summary()
```

```python
Model.add(Dense(10, activation = 'softmax'))
```

```python
Model.summary()
```

```python
Model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
Model.fit(X_train, y_train, epochs = 5, batch_size = 128)
```

```python
X_test.shape
```

```python
test_loss, test_accuracy = Model.evaluate(X_test, y_test)
```

```python
Prediction_X = Model.predict(X_test)
```

```python
Prediction_X
```

```python
Prediction_class = Model.predict_classes(X_test)
```

```python
Prediction_class
```

```python
Prediction_class = Prediction_class.reshape(-1,1)
```

```python
test_Y = test_Y.reshape(-1,1)
```

```python
from sklearn.metrics import confusion_matrix
MNIST_Confusion = confusion_matrix(test_Y, Prediction_class)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

fig, axe = plt.subplots(1,figsize = (6,4))
sn.heatmap(MNIST_Confusion, annot = True, ax = axe, cmap = 'Blues')

plt.show()
```
