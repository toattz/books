```python
def test_function03(s,func):
    for item in s:
        print(func(item[0],item[1]))
```

```python
s = [(1,2),(3,4),(5,6)]
```

```python
def Area(length,width):
    return length*width
```

```python
test_function03(s,Area)
```

```python
test_function03(s,lambda length,width:length*width)
```

```python
Area02 = lambda length,width:length*width
```

```python
test_function03(s,Area02)
```

```python
def test_function04(List01, List02):
    List01.append(3)
    List02 = [1, 2, 3]
    print(f'List01 = {List01} and List02 = {List02}')
```

```python
My_List01 = [1, 2]
My_List02 = [4, 5, 6]
test_function04(My_List01, My_List02)
print(f'My_List01 = {My_List01} and My_List02 = {My_List02}')
```

```python
def test_function05(List01, List02):
    List01 = List01[:]
    List01.append(3)
    List02 = [1, 2, 3]
    print(f'List01 = {List01} and List02 = {List02}')
```

```python
My_List01 = [1, 2]
My_List02 = [4, 5, 6]
test_function05(My_List01, My_List02)
print(f'My_List01 = {My_List01} and My_List02 = {My_List02}')
```
