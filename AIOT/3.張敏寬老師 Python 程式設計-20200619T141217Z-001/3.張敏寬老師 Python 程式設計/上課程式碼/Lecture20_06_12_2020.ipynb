```python
import matplotlib.pyplot as plt

Target_name = {0: 'airplane', 1: 'auto', 2: 'bird', 3: 'cat', 4: 'deer',\
5: 'dog', 6: 'frog', 7: 'horse', 8: 'ship', 9: 'truck'}

plt.figure(figsize = (10,8))

for i in range(40):
    plt.subplot(4, 10, i+1)
    plt.imshow(train_X[i])
    plt.title(f'{Target_name[train_y.ravel()[i]]}')
    plt.axis('off')

plt.subplots_adjust(wspace = 0.2, hspace = 0.2)
plt.show()
```

```python
X_train = train_X.astype('float')/255
X_test = test_X.astype('float')/255
```

```python
train_X[0][0][0]
```

```python
X_train[0][0][0]
```

```python
print(train_y[:5])
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(train_y)
y_test = to_categorical(test_y)
```

```python
y_train[:5]
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

CNN_Model = Sequential()
CNN_Model.add(Conv2D(16, (3,3), activation = 'relu', input_shape =(32,32,3)))
CNN_Model.add(MaxPooling2D(2,2))
CNN_Model.add(Conv2D(16, (3,3), activation = 'relu'))
CNN_Model.add(MaxPooling2D(2,2))
CNN_Model.add(Flatten())
CNN_Model.add(Dense(32, activation ='relu'))
CNN_Model.add(Dense(10, activation = 'softmax'))
CNN_Model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
CNN_Model.fit(X_train, y_train, epochs = 5, batch_size = 128, validation_split = 0.1, verbose = 0)
test_loss, test_acc = CNN_Model.evaluate(X_test, y_test)
print(test_loss, test_acc, sep = ',')
```
